# Mini-CAS

## Sintaxis

La utilidad CAS a implementar debe usar la sintaxis matematica de [python](https://docs.python.org/3/reference/expressions.html) o de [scheme](https://www.gnu.org/software/mit-scheme/documentation/mit-scheme-ref/Numerical-operations.html) para las expresiones matematicas, y estas deberán ser incluidas como cadena de caracteres.  

    expresion = 'f = 2 * x**3 + (x - a)**2 / 2'

o

    expresion = '= f + * 2 ** x 3 / ** - x a 2 2'
    
Por conveniencia, si usa la sintaxis de scheme, es conveniente usar agrupación para notar más claramente la aplicación de los operadores aunque no es necesario dada la definición de la aplicación de los operadores (unarios y binarios).  

    expresion = '(= f (+ (* 2 (** x 3)) (/ (** (- x a) 2) 2)))'
    
Por conveniencia se reemplazaron las funciones `define` y `expt` por sus equivalentes python `=` y `**`.
    
## Caracteristicas de la implementación


+ Todo valor númerico debe representarse de forma exacta.  
    + Se deben soportar los números enteros.  
    + Se debe soportar los números reales de forma decimal (con separador anglosajon).  
    + Se deben soportar los números reales de forma fraccionaria.  
+ Toda variable se identifica por un caracter alfabetico individual.  
    + Se debe soportar como variable para evaluación la `x`.  
    + Se debe soportar la asignación de expresiones algebraicas a una variable.  
    + Toda asignación que contenga `x` convierte la variable en una función.  
    + Se debe soportar como coeficiente toda variable que no ha sido asignada y no sea `x`.  
    + Si un coeficiente es usado para asignación, debe sustituirse en todas las expresiones por la expresión asignada (comportamiento regresivo y progresivo).  
+ Toda función por defecto se expresa sin expandir o evaluar.  
    + Se debe soportar el uso de la función `sqrt`.  
        + Se debe soportar la evaluación exacta en $0$, $1$ y $4$.  
        + Admite como argumento expresiones algebraicas.  
+ Aritmetica real exacta para los operadores `+`, `-`, `*` y `**` (exponente entero).  
    + Operaciones que involucran fraccionarios se expresan como fraccionarios los resultados.  
    + Por defecto, el resultado de la potencia no se expande.  
    + La división y cantidades fraccionarias solo se expresan como la fracción.  
        + Si el denominador es $1$ se debe simplificar por defecto al real o entero del numerador.  
        + Si el numerador es $0$ se debe simplificar por defecto al entero $0$.  
+ Operaciones algebraicas / polinomicas.  
    + Suma y resta algebraica.  
    + Producto algebraico (soporte de exponentes enteros).  
    + Potenciación (soporte de exponentes enteros). Por defecto no se expande.  
    + Las operaciones aritmeticas deben seguir por defecto su comportamiento expuesto.  
    + Por defecto la propiedad distributiva no se aplica.  
+ Funciones para el usuario: Se deben crear funciones para los siguientes casos
    + `polinomio`: Valida si la expresión es polinomica en `x`. Debe reconocer como polinomio la forma tradicional y formas factorizadas. Debe funcionar en expresiones con y sin asignación.  
    + `simplificar`: Realiza las simplificaciones dadas por los comportamientos por defecto solo si es polinomica. Debe funcionar en expresion con y sin asignación.  
    + `registrar`: Agrega la expresión al registro de la utilidad CAS solo si es polinomica. El registro es la forma mediante la cual se tendrá almacenadas las expresiones con asignación. Si se registra una variable existente en las expresiones registradas debe sustituirse. Una expresión almacenada debe verificar si ya existe una sustitución para sus variables.  
    + `evaluar`: Evalua un polinomio en una expresión numérica solo si sus coeficientes son numéricos. La evaluación ejecuta los comportamientos por defecto. Debe funcionar en expresiones registradas y no registradas.  
    + `raices`: Dado un polinomio de grado uno o dos expresado en cualquier forma, o un polinomio factorizado por producto de factores, debe indicar todas sus raices de forma exacta. Debe funcionar con expresiones registradas y sin registrar.  
    + `expandir`: Expande los casos especiales de evaluaciones exactas de la raíz y la división, al igual que expande las potencias y aplica la propiedad distributiva. Si existen variables asignadas, reemplaza las expresiones correspondientes y evalua las funciones.  
    + `recuperar`: Recupera una expresión almacenada en el registro por medio del nombre de la variable.
